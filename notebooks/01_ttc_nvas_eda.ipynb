# 01 - TTC NVAS EDA (Milestone 1)

Assumptions:
- NVAS access is GTFS-Realtime from TTC BusTime endpoints.
- For this first milestone, we use a small CSV sample in `data/raw/` for pandas-friendly EDA.


```python
from pathlib import Path
import pandas as pd

raw_path = Path('../data/raw/ttc_nvas_vehicle_positions_sample.csv')
print('File path used:', raw_path.resolve())
df = pd.read_csv(raw_path)
df.head()
```

```python
print('Shape:', df.shape)
print('Columns:', list(df.columns))
```

```python
null_summary = df.isna().sum().sort_values(ascending=False)
duplicate_count = int(df.duplicated().sum())
print('Null summary:')
display(null_summary)
print('Duplicate summary:')
print(duplicate_count)
```

```python
df['timestamp_parsed_utc'] = pd.to_datetime(df['timestamp_utc'], errors='coerce', utc=True)
print('Parsed timestamp nulls:', int(df['timestamp_parsed_utc'].isna().sum()))
df[['timestamp_utc','timestamp_parsed_utc']].head()
```
